Что делает команда python -m venv venv?
Создает виртуальное окружение Python в папке venv. Оно изолирует зависимости для проекта, предотвращая конфликты с другими проектами.

1.1 Что делает каждая команда в списке ниже?

pip list                       # 0 Выводит список установленных пакетов в текущем окружении.
pip freeze > requirements.txt  # 1 Создает файл requirements.txt со списком установленных пакетов и их версий, пригодный для воспроизведения окружения
pip install -r requirements.txt # 2 Устанавливает пакеты, перечисленные в файле requirements.txt.

2. Что делает каждая команда в списке ниже?

conda env list                            # 1 Выводит список существующих окружений conda.
conda create -n env_name python=3.5       # 2 Создает новое окружение conda с именем env_name и версией Python 3.5.
conda env update -n env_name -f file.yml  # 3 Обновляет существующее окружение conda env_name на основе зависимостей, указанных в файле file.yml.
source activate env_name                  # 4 Активирует окружение conda с именем env_name (в более новых версиях conda рекомендуется conda activate env_name).
source deactivate                         # 5 Деактивирует текущее окружение conda (в более новых версиях conda рекомендуется conda deactivate).

conda clean -a                            # 6 Удаляет неиспользуемые пакеты, кэши и временные файлы conda, освобождая место на диске.

8. Где по умолчанию больше пакетов venv/pip или conda? и почему дата сайнинисты используют conda?
venv/pip: Ориентированы в основном на пакеты Python.
conda: Предназначена для управления зависимостями для проектов с использованием различных языков программирования (Python, R, C++, и т.д.) и имеет встроенную поддержку бинарных зависимостей. Это особенно важно для data science, так как многие библиотеки (например, NumPy, SciPy) содержат код, написанный на C/C++, и conda упрощает установку этих библиотек с необходимыми зависимостями. Conda также проще в управлении версиями системных библиотек (например, CUDA для GPU).
Поэтому conda выбирают, чтобы обеспечить воспроизводимость среды и простоту установки пакетов, критичных для научных вычислений и анализа данных.
