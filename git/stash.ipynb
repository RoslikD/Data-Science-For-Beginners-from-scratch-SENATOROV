Что делает команда git stash?
Команда git stash временно сохраняет (прячет) текущие незакоммиченные изменения в рабочем каталоге и индексе (staging area).

Как просмотреть список всех сохранённых изменений (стэшей)?
git stash list

Какая команда применяется для использования верхнего стэша?
git stash apply

Как применить конкретный стэш по его номеру?
git stash apply stash@{номер}

Чем отличается команда git stash apply от git stash pop?
git stash apply: Применяет стэш, но оставляет его в стэш-листе.
git stash pop: Применяет стэш и удаляет его из стэш-листа.

Что делает команда git stash drop?
Удаляет стэш с указанным номером из стэш-листа. 

Как полностью очистить все сохранённые стэши?
git stash clear

В каких случаях удобно использовать git stash?
Переключение веток
Обновление ветки
Тестирование
Разделение изменений

Что произойдёт, если выполнить git stash pop, но в проекте есть конфликтующие изменения?
Git попытается применить изменения из стэша, но если они конфликтуют с текущими изменениями в вашем рабочем каталоге, возникнет конфликт слияния (merge conflict).

Можно ли восстановить удалённый стэш после выполнения git stash drop?
Восстановить удаленный стэш возможно (но гарантий нет)

Что делает команда git stash save "NAME_STASH"
создает новый стэш с заданным именем NAME_STASH

Что делает команда git stash apply "NUMBER_STASH"
Вместо NUMBER_STASH нужно указать имя стэша. 

Что делает команда git stash pop "NUMBER_STASH"
применяет и удаляет стэш с номером NUMBER_STASH из списка.
