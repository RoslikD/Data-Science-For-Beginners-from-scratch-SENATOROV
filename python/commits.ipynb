"""Task commits."""

Опишите своими словами назначение каждого из этих типов коммитов:
feat, fix, docs, style, refactor, test, build, ci, perf, chore.

feat (feature - функциональность): Этот тип коммита используется для добавления новой функциональности в код.
fix (исправление): Этот тип коммита используется для исправления ошибок или багов в коде.
docs (документация): Этот тип коммита предназначен для изменений, связанных с документацией.
style (стиль): Этот тип коммита относится к изменениям в стиле кода, форматированию и других аспектах, которые не влияют на функциональность
refactor (рефакторинг): Этот тип коммита используется для рефакторинга кода.
test (тесты): Этот тип коммита используется для добавления, изменения или удаления тестов.
build (сборка): Этот тип коммита относится к изменениям в системе сборки или зависимостях.
ci (continuous integration - непрерывная интеграция): Этот тип коммита связан с изменениями в настройках системы непрерывной интеграции (CI). 
perf (performance - производительность): Этот тип коммита предназначен для изменений, которые улучшают производительность кода.
chore (рутина, хозяйственные задачи): Этот тип коммита используется для изменений, которые не попадают ни в одну из вышеперечисленных категорий.


Представьте, что вы исправили баг в функции, которая некорректно округляет числа. Сделайте фиктивный коммит и напишите для него сообщение в соответствии с Conventional Commits (используя тип fix).
git commit -m "fix: Correctly round numbers in calculation function"

Добавление новой функциональности:
Допустим, вы реализовали новую функцию generateReport в проекте. Сделайте фиктивный коммит с типом feat, отражающий добавление этой функциональности
git commit -m "feat: Implement generateReport function to create project reports"

Модификация формата кода или стилей:
Представьте, что вы поправили отступы и форматирование во всём проекте, не меняя логики кода. Сделайте фиктивный коммит с типом style
git commit -m "style: Format code to adhere to project coding standards"

Документация и тестирование:

Сделайте фиктивный коммит с типом docs, добавляющий или улучшающий документацию для вашей новой функции.
git commit -m "docs: Add JSDoc comments for the generateReport function"

Сделайте фиктивный коммит с типом test, добавляющий тесты для этой же функции.
git commit -m "test: Add unit tests for the generateReport function"
